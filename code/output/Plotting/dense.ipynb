```python
import csv
import numpy as np

list1 = []
list2 = []
with open('./edges.csv',newline='') as f:
    Edges = csv.reader(f, delimiter=',')
    for x in Edges:
        list1.append(np.array(x).astype(np.float64))
        
list1 = np.array(list1)
with open('./vertices.csv',newline='') as f:
    Edges = csv.reader(f, delimiter=',')
    for x in Edges:
        list2.append(np.array(x).astype(np.float64))
list2 = np.array(list2)
print(list1[:,:])
print()
print(list2)
```

```python
v500 = list1[np.where(list1[:,0]==500)]

x_axis = v500[:,1]
build = v500[:, 2]
oracle_const = v500[:,3]
oracle_query = v500[:,4]
fib_const = v500[:,5]
fib_query = v500[:,6]
fw = v500[:,7]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8), dpi=100)
plt.scatter(x_axis, oracle_const, color='hotpink', label='Oracle Construction')
plt.scatter( x_axis, fib_const, color='green', label='Fibonacci Oracle Construction')
plt.scatter(x_axis, fw, color='orange', label='Floyd-Warshall')

plt.xlabel('No. of edges')
plt.ylabel('Runtime in ms for n=500')
plt.title('Dense Graphs\nRuntime vs No. of edges for constant number of vertices')
plt.legend()
plt.show()
```

```python
oc2 = oracle_const + 500**2*oracle_query
fc2 = fib_const + 500**2*fib_query
plt.figure(figsize=(10, 8), dpi=100)
plt.scatter(x_axis, oc2, color='hotpink', label='Oracle Construction + All queries')
plt.scatter( x_axis, fc2, color='green', label='Fibonacci Oracle Construction + All Queries')
plt.scatter(x_axis, fw, color='orange', label='Floyd-Warshall')

plt.xlabel('No. of edges')
plt.ylabel('Runtime in ms for n=500')
plt.title('Dense Graphs\nRuntime vs No. of edges for constant number of vertices')
print(oc2-oracle_const)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 8), dpi=100)
plt.scatter(x_axis, oracle_query, color='hotpink', label='Oracle Queries')
plt.scatter( x_axis, fib_query, color='green', label='Fibonacci Oracle Queries')

plt.xlabel('No. of edges')
plt.ylabel('Runtime in ms for n=500')
plt.title('Dense Graphs\nQuery time vs No. of edges for constant number of vertices')
plt.legend()
plt.show()
```
